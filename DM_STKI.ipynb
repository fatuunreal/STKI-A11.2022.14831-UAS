### Memakai kolom full_text atau Tweet


```python
# prompt: hapus seluruh kolom kecuali 'full_text' dan ubah nama kolom menjadi 'Tweet'

# Keep only the 'full_text' column and rename it to 'Tweet'
df = df[['full_text']].rename(columns={'full_text': 'Tweet'})

# Display the modified DataFrame
df.head(5)
```

```python
# prompt: Convert to Lowercase, Handle Special Characters and Numbers: Consider retaining emojis and hashtags as they may carry.
# dan mengganti semua username ke @user

import pandas as pd
import re

# Assuming df is your DataFrame with the 'Tweet' column
# ... (your existing code to load the DataFrame) ...

def clean_text(text):
    # Lowercase conversion
    text = text.lower()

    # Replace usernames with @user
    text = re.sub(r'@\w+', '@user', text)

    #Handle special characters and numbers (retain emojis and hashtags)
    #This regex keeps alphanumeric characters, whitespace, emojis, and hashtags
    #You might need to adjust the regex based on the specific characters you want to keep/remove.
    text = re.sub(r'[^a-zA-Z0-9\s#@\U00010000-\U0010ffff]', '', text)

    return text


df['cleaned_tweet'] = df['Tweet'].apply(clean_text)

# Display the DataFrame with the cleaned tweets
df.head(5)
```

# Tokenisasi

```python
# token
import nltk
#Download punkt_tab sebagai koreksi dari error yang diberikan
nltk.download('punkt_tab')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
def word_tokenize_wrapper(text):
  return word_tokenize(text)
```

```python
#Menambahkan kolom tokens pada df
df['tokens'] = df['cleaned_tweet'].apply(word_tokenize_wrapper)
```

```python
# prompt: lakukan stopword removal

import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download necessary NLTK data (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# ... (your existing code to load and clean the DataFrame) ...

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('indonesian'))  # Use Indonesian stopwords
    filtered_tokens = [token for token in tokens if token not in stop_words and token.isalnum()] # remove non-alphanumeric tokens
    return filtered_tokens

# Apply stopword removal to the 'tokens' column
df['tokens_no_stopwords'] = df['tokens'].apply(remove_stopwords)

df.head(10)
```

# Stemming dengan Sastrawi

```python
# prompt: LAKUKAN STEMMING MENGGUNAKAN LIBRARY SASTRAWI

!pip install Sastrawi

import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# ... (your existing code to load the DataFrame and perform cleaning) ...

# Create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming_wrapper(text):
  return stemmer.stem(text)

#Menambahkan kolom stemmed_tokens pada df
df['stemmed_tokens'] = df['tokens'].apply(lambda x: [stemmer.stem(token) for token in x])
df.head(10)
```

```python
normalized_word = pd.read_excel("kamus perbaikan kata.xlsx")

normalized_word_dict = {}

for index, row in normalized_word.iterrows():
    if row[0] not in normalized_word_dict:
        normalized_word_dict[row[0]] = row[1]

def normalized_term(document):
    return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

df['tokens_perbaikan'] = df['stemmed_tokens'].apply(normalized_term)
df.head()
```

```python
# prompt: Labeling the Data
# Use Pre-trained Model, bahasa indonesia

!pip install transformers

import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Load a pre-trained Indonesian sentiment analysis model
model_name = "indobenchmark/indobert-base-p2"  # Or another suitable model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Assuming 'df' is your DataFrame with the 'stemmed_tokens' column
# ... (your existing code to load and process the DataFrame) ...

def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        predicted_class = torch.argmax(logits, dim=-1).item()
    return predicted_class

# Apply the sentiment prediction function to the 'stemmed_tokens' column
# Join the stemmed tokens back into a string for prediction
df['predicted_label'] = df['stemmed_tokens'].apply(lambda tokens: predict_sentiment(" ".join(tokens)))

df.head(10)
```
